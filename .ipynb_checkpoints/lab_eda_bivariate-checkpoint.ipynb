```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = '../../../../week5/amz_uk_dataset.csv'
df = pd.read_csv(data)
df.min()
```

```python
df[df.columns[df.isnull().mean() < 0.8]]
```

```python
potential_categorical_from_numerical = df.select_dtypes('number').loc[:, df.select_dtypes('number').nunique() < 20]
```

```python
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])
crosstab_result_normalized = pd.crosstab(df['category'], df['isBestSeller'], normalize= 'all')
crosstab_result_normalized = crosstab_result_normalized.sort_values(by=True, ascending=False)
crosstab_result_normalized
```

- The categories where a BestSeller product is more prevalent are Grocery, Health & Personal Care, and Baby.

```python
from scipy.stats import chi2_contingency
chi2_statistic, chi2_p_value, _, _ = chi2_contingency(crosstab_result)

chi2_statistic, chi2_p_value
```

Given our P-value is higher than 0.05, we can determine there is no relationship between the category and isBestSeller column.


```python
crosstab_result = crosstab_result.astype(int)
from scipy.stats.contingency import association
association(crosstab_result, method="cramer")
```

To conduct the Cramer V test, we need to convert the crosstab results into integers. To do that, they are automatically rounded to 0, given the low value they have, and therefore the result is 'nan', as we cannot divide by 0. In any case, the Cramer V test is conducted to calculate the strenth of the relationship between two categorical columns, and we have already determined that there is no relationship between these two columns based on the P-value calculated in the Chi-2 test.

```python
top_crosstab_result = crosstab_result_normalized.iloc[:6, :]
top_crosstab_result.plot(kind='bar', stacked=True)
```

The graph above is supposed to show the percentage of products that are best sellers for the categories that have higher percentages of BestSellers. The percentage is practically invisible, which means there is no category that we could highlight based on the number of BestSellers it produces. 

```python
#removing outliers in product prices

Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + (1.5*IQR)
lower_bound = Q1 - (1.5*IQR)

df_wo_outliers = df[(df['price'] < upper_bound) & (df['price'] > lower_bound)]
df_wo_outliers.sort_values(by=['price', 'title'], ascending=False) 
```

```python
print(f'the upper bound is {upper_bound}, and the lower bound is: {lower_bound}') 
```

```python
top_10_categories = df_wo_outliers['category'].value_counts().head(10).index
new_df = df_wo_outliers[df_wo_outliers['category'].isin(top_10_categories)]

sns.violinplot(data = new_df , x= 'category', y= 'price', palette = 'coolwarm', hue= 'category', legend=False)

plt.xticks(rotation=90)  # Rotate x-axis labels to 90 degrees
plt.show()
```

Filtering the top 20 categories still does not provide a clear representation, so I decided to filter out the top 10 categories. Now we can see all median prices across categories are under $20, and we can also see prices are very similarly distributed across these categories.

```python
new_df.groupby('category')['price'].agg('mean').sort_values(ascending=False)
```

```python
sns.barplot(data = new_df , x= 'category', y= 'price', palette = 'coolwarm', hue= 'category', legend=False)
plt.xticks(rotation=90)  # Rotate x-axis labels to 90 degrees
plt.show()
```

```python
sns.barplot(data = df_wo_outliers , x= 'category', y= 'price', palette = 'coolwarm', hue= 'category', legend=False)
plt.xticks(rotation=90)  # Rotate x-axis labels to 90 degrees
plt.show()
```

If we don't filter out the top 10 categories, it is very difficult to see the category with the highest mean given the amount of categories that we are working with, as shown above.

```python
sns.boxplot(data = new_df , x= 'category', y= 'stars', palette = 'coolwarm', hue= 'category', legend=False)
plt.xticks(rotation=90)  # Rotate x-axis labels to 90 degrees
plt.show()
```

According to the graph above, the highest median ratings are for hair care, skin care, and Bath & Body, as demonstrated with the calculation below.

```python
new_df.groupby('category')['stars'].median().sort_values(ascending=False)
```

```python
#Part 3
correlation = df_wo_outliers['price'].corr(df_wo_outliers['stars'])
correlation
```

By calculating the correlation between price and ratings, we can see that they have a negative correlation, meaning when price goes up, ratings go down.
However, it is a very weak correlation, given the range is -1 to 1. As such, we can determine there is not a strong dependency between the two variables.

```python
sns.scatterplot(data = new_df, x= 'stars', y= 'price', color= 'skyblue', alpha= 0.5);
```

The graph above shows the lack of dependency between these two variables. All star ratings have data points across the whole price range. In other words, we do not see that when one variable changes, the other variable responds to the change.

```python
df_numerical = df_wo_outliers.select_dtypes('number')
correlation_matrix = df_numerical.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

With the heatmap we can see there is barely any correlation between any of the numerical columns, meaning all columns are independent of each other. The strongest relationship that we can see is 0.11, which is still very low but shows some positive relationship between, for example, reviews and bought in last month, which implies that if the product gets bought more often during the month, it will obtain more reviews.
